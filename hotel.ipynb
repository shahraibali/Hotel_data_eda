# Data Analysis and Visuizations

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print('cancelled_perc')

plt.title('Reservation state count')

plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('Reservation status in different hotells',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']# filter data
resort_hotel['is_canceled'].value_counts(normalize=True)# check persentage
```

```python
City_hotel=df[df['hotel']=='City Hotel']
City_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel=City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Dailt Rate in city and Resort Hotel',fontsize=30)

plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(City_hotel.index,City_hotel['adr'],label='City Hotel')

plt.legend(fontsize=20)
plt.show()
```

```python
# Convert the 'reservation_status_date' column to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

# Extract the month from the 'reservation_status_date' column
df['month'] = df['reservation_status_date'].dt.month
```

```python
#df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
axl=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')

lines,labels=axl.get_legend_handles_labels()

axl.legend(lines,labels,fontsize=15)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend('not canceled','canceled')

plt.show()
```
